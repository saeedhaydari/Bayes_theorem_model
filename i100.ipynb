```python
# با استفاده از مدل بیز ساده، بگویید کدام مدل دستگاه را به شخصی با اولویت های زیر پیشنهاد می دهید؟
# شخصی که اولیتش سلامتی ، و در سطح متعادل ورزش می کند، به صورت نسبی مشتاق به تمرین کردن است 
#و  در  استفاده از دستگاه های تکنولوژی راحت است
# باید احتمال های زیر را محاسبه کنیم هر کدام بیشتر بود آن دستگاه را به مشتری پیشنهاد بکنیم
# P(i100|health, moderate_exercise, moderate_motivete, yes) = P(health|i100)*P(moderate_exercise|i100)*P(moderate_motivete|i100)*P(yes|i100)*p(i100)
# P(i500|health, moderate_exercise, moderate_motivete, yes)= P(health|i500)*P(moderate_exercise|i500)*P(moderate_motivete|i500)*P(yes|i500)*p(i500)
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('i100.csv')
```

```python
data.head(15)
```

```python
# evidence = health, moderate_exercise, miderate_motivate, 
# calculate P(i100|evidence)

#  calculate P(i100)
i100 = data[data["model"] == 'i100']  # count the number of i100 among dataset
p_i100= len(i100)/len(data)           # calculate the probabilty of i100


# calculate P(health|i100)
health_i100 = data[(data["main_interest"] == 'health') & (data["model"] == 'i100')]     # Count health intersect with i100
p_health_i100 = len(health_i100)/len(i100)                                              # calculate P(health|i100) 


# calculate P(moderate_exercise|i100)
moderate_exercise_i100= data[(data["current_exercise_level"] == 'moderate') & (data["model"] == 'i100')]  # Count current_exercise_level intersect with i100
p_moderate_exercise_i100 = len(moderate_exercise_i100)/len(i100)          #calculate P(moderate_exercise|i100)
 

# calculate P(moderate_motivete|i100)
moderate_motivete_i100= data[(data["how_motivated"] == 'moderate') & (data["model"] == 'i100')]  # Count how_motivated intersect with i100
p_moderate_motivete_i100 = len(moderate_motivete_i100)/len(i100)            #calculate P(moderate_motivate|i100)
 

# calculate P(yes|i100)
yes_i100= data[(data["comfortabl_with_tech_device"] == 'yes') & (data["model"] == 'i100')]  # Count comfortabl_with_tech_device interset with i100
p_yes_i100 = len(yes_i100)/len(i100)            #calculate P(yes|i100)
 

# calculate P(i100|evidence) = P(health|i100) * P(moderate_exercise|i100) * P(moderate_motivete|i100) * P(yes|i100) * p(i100)
P_i100_evidence = (p_health_i100) * (p_moderate_exercise_i100) * (p_moderate_motivete_i100) * (p_yes_i100) * (p_i100)
P_i100_evidence
```

```python
# evidence = health, moderate_exercise, miderate_motivate, 
# calculate P(i500|evidence)


#  calculate P(i500)
i500 = data[data["model"] == 'i500']
p_i500= len(data[data["model"] == 'i500'])/len(data)    # calculate the probabilty of i100


# calculate P(health|i500)
health_i500 = data[(data["main_interest"] == 'health') & (data["model"] == 'i500')]     # Count health intersect with i500
p_health_i500 = len(health_i500)/len(i500)            #  calculate P(health|i500) 


# calculate P(moderate_exercise|i500)
moderate_exercise_i500= data[(data["current_exercise_level"] == 'moderate') & (data["model"] == 'i500')]  # Count current_exercise_level intersect with i500
p_moderate_exercise_i500 = len(moderate_exercise_i500)/len(i500)            #  c calculate P(moderate_exercise|i500)
 

# calculate P(moderate_motivete|i500)
moderate_motivete_i500= data[(data["how_motivated"] == 'moderate') & (data["model"] == 'i500')]  # Count how_motivated intersect with i500
p_moderate_motivete_i500 = len(moderate_motivete_i500)/len(i500)            #   calculate P(moderate_motivate|i500)
 

# calculate P(yes|i500)
yes_i500= data[(data["comfortabl_with_tech_device"] == 'yes') & (data["model"] == 'i500')]  # Count comfortabl_with_tech_device intersect with i500
p_yes_i500 = len(yes_i500)/len(i500)            #   calculate P(yes|i500)
 
# evidence = health, moderate_exercise, miderate_motivate, 
# calculate P(i500|evidence) = P(health|i500) * P(moderate_exercise|i500) * P(moderate_motivete|i500) * P(yes|i500) * p(i500)
P_i500_evidence = (p_health_i500) * (p_moderate_exercise_i500) * (p_moderate_motivete_i500) * (p_yes_i500) * (p_i500)
P_i500_evidence
```

```python
print ('probablity of I100 is:',"{:.5f}".format(P_i500_evidence))
print ('probablity of I500 is:',"{:.5f}".format(P_i100_evidence))
```
